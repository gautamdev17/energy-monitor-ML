```python
#importing all essential libraries
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
#including training data
#each row: [month,temperature,humidity%]
#both temp and humidity are medians over the month
X = np.array([
    [1, 25.1, 70.3],
    [2, 26.8, 65.7],
    [3, 30.2, 60.1],
    [4, 32.5, 58.6],
    [5, 34.0, 64.3],
    [6, 33.5, 66.0]
])
#corresponding total kwh consumed in those months
y = np.array([120.5, 114.3, 132.0, 140.8, 148.9, 150.2])
```

```python
#now lets train the model using linear regression
model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)
months = X[:, 0]

plt.scatter(months, y, color='blue', label='Actual')
plt.plot(months, y_pred, color='red', linestyle='--', label='Predicted')
plt.xlabel("Month")
plt.ylabel("Power Consumed (kWh)")
plt.title("Linear Regression: Month vs Power")
plt.legend()
plt.grid(True)
plt.show()
```

```python
r2 = r2_score(y, y_pred)
print(f"R² Score       : {r2:.4f}")
```

```python
july_input = np.array([[7, 33.0, 68.0]])  # Month 7, Temp 33.0°C, Humidity 68%
july_kwh = model.predict(july_input)[0]

rate_per_kwh = 6.50  # ₹6.50 per unit
july_bill = july_kwh * rate_per_kwh

print(f"🔋 Predicted July usage : {july_kwh:.2f} kWh")
print(f"💸 Estimated July bill   : ₹{july_bill:.2f}")
```
